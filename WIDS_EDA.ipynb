```python
#output of this cell will give the graph between match vs number of ball were left when chasing team won the match
#more balls implies greater victory

df1 = data_matches[["id", "win_by_runs", "win_by_wickets"]]
df2 = data_deliveries[["match_id", "inning", "over", "ball"]]
df1.columns = ["match_id", "win_by_runs", "win-by_wickets"]
df = pd.merge(df1, df2)
df = df[df["win_by_runs"] == 0]
ball_dict = {}
ball_list = []

for id in df["match_id"].unique():
  df3 = df[df["match_id"] == id]
  df4 = df3[df3["inning"] == 2]
  ball_dict[list(120 - ((df4.tail(1)["over"]-1)*6 + (df4.tail(1)["ball"])))[0]] = id
  ball_list.append(list(120 - ((df4.tail(1)["over"]-1)*6 + (df4.tail(1)["ball"])))[0])

id_list = []
for balls in sorted(ball_list, reverse = True)[:10]:
  id_list.append(ball_dict.get(balls))

team_list = []
for id in id_list:
  d = data_matches[data_matches["id"] == id]
  team_1 = list(d["team1"])[0]
  team_2 = list(d["team2"])[0]
  year = list(d["season"])[0]
  team_list.append(f"{team_1} vs {team_2} ({year})")

trace = go.Bar(
      x = sorted(ball_list, reverse = True)[:10], 
      y = team_list,
      orientation = "h"
)

layout = go.Layout(
    title = "balls left when a team won",
    xaxis = dict(title = "Balls"),
    yaxis = dict(title = "Teams")
)

plot = go.Figure(
    data = [trace],
    layout = layout
)

plot.show()
```

## Compairing two given batsman

```python
def get_barh(batsman1, batsman2):
  # This function will give barh graph of man of the match between batsman1 and batsman2
  y = [batsman1, batsman2]
  if (batsman1 and batsman2) in data_matches["player_of_match"].value_counts().index:
    width = [data_matches["player_of_match"].value_counts()[batsman1], data_matches["player_of_match"].value_counts()[batsman2]]
  elif (batsman1 in data_matches["player_of_match"].value_counts().index) and (batsman2 not in data_matches["player_of_match"].value_counts().index):
    print(f"either {batsman2} hasn't won any man of the match award or {batsman2} is incorrect name of player")
    width = [data_matches["player_of_match"].value_counts()[batsman1], 0]
  elif (batsman2 in data_matches["player_of_match"].value_counts().index) and (batsman1 not in data_matches["player_of_match"].value_counts().index):
    print(f"either {batsman1} hasn't won any man of the match award or {batsman1} is incorrect name of player")
    width = [0, data_matches["player_of_match"].value_counts()[batsman2]]
  else:
    print(f"either {batsman1} and {batsman2} hasn't won any man of the match award or {batsman1} and {batsman2} are incorrect names of players")
    width = [0,0]
  fig = go.Figure(go.Bar(x = width, y = y, orientation = "h"))
  trace = go.Bar(
      x = width,
      y = y,
      orientation = "h"
  )

  layout = go.Layout(
      title = f"{batsman1} vs {batsman2} man of the match award won",
      xaxis = dict(title = "No. of times man of the match award won"),
      yaxis = dict(title = "batsmen")
  )

  fig = go.Figure(
      data = [trace],
      layout = layout
  )

  fig.show()
```

```python
def compair(batsman1, batsman2):
  #this fn gives a side by side bar graph of total runs scored by two batsmen
  batsman1_score_list = []
  batsman2_score_list = []

  df1 = data_deliveries[["match_id","batsman", "batsman_runs"]]
  df1.columns = ["id", "batsman", "batsman_runs"]
  df2 = data_matches[["id", "season"]]
  df3 = pd.merge(df1, df2)
  df4 = df3.drop(columns="id")

  for year in sorted(df4[df4["batsman"] == batsman1]["season"].unique()):
    score = df4[df4["batsman"] == batsman1][df4[df4["batsman"] == batsman1]["season"]==year]["batsman_runs"].sum()
    batsman1_score_list.append(score)

  for year in sorted(df4[df4["batsman"] == batsman2]["season"].unique()): 
    score = df4[df4["batsman"] == batsman2][df4[df4["batsman"] == batsman2]["season"]==year]["batsman_runs"].sum()
    batsman2_score_list.append(score)    


  data = []

  trace1 = go.Bar(
      x = sorted(df4[df4["batsman"] == batsman1]["season"].unique()),
      y = batsman1_score_list,
      name = batsman1
  )

  trace2 = go.Bar(
      x = sorted(df4[df4["batsman"] == batsman1]["season"].unique()),
      y = batsman2_score_list,
      name = batsman2
  )  

  data = [trace1, trace2]

  layout = go.Layout(
      title = f"{batsman1} vs {batsman2} season wise total runs",
      xaxis = dict(title = "season",
                   tickmode = "linear"),
      yaxis = dict(title = "total runs made") 
  )

  fig = go.Figure(
      data = data,
      layout = layout
  )

  fig.show()
```

```python
compair("CH Gayle", "DA Warner")
```

```python
def compaire_centuries_half_centuries(batsman1, batsman2, compair_half_centuries = True, compair_centuries = True):
  #this function gives barh graph of total no of centuries and half centuries by teo batsmen
  df1 = data_deliveries[["match_id","batsman", "batsman_runs"]]
  df2 = df1[df1["batsman"] == batsman1]
  df3 = df1[df1["batsman"] == batsman2]
  runs_list_batsman1 = []
  runs_list_batsman2 = []

  for id in df2["match_id"].unique():
    score = df2[df2["match_id"] == id]["batsman_runs"].sum()
    runs_list_batsman1.append(score)

  for id in df3["match_id"].unique():
    score = df3[df3["match_id"] == id]["batsman_runs"].sum()
    runs_list_batsman2.append(score)

  runs_list_batsman1 = np.array(runs_list_batsman1)
  runs_list_batsman2 = np.array(runs_list_batsman2)

  if compair_half_centuries:
    width_half_centuries = [len(np.where(runs_list_batsman1 >= 50)[0]), len(np.where(runs_list_batsman2 >= 50)[0])]
    y_half_centuries = [batsman1, batsman2]

    trace = go.Bar(
        x = width_half_centuries, 
        y = y_half_centuries, 
        orientation = "h"
    )

    Layout = go.Layout(
    title = "batsmen's performance",
    xaxis = dict(title = "No. of half centuries"),
    yaxis = dict(title = "Batsmen")
    )

    fig = go.Figure(
        data = [trace],
        layout = Layout
    )

    fig.show()

  if compair_centuries:
    width_centuries = [len(np.where(runs_list_batsman1 >= 100)[0]), len(np.where(runs_list_batsman2 >= 100)[0])]
    y_centuries = [batsman1, batsman2]

    trace = go.Bar(
        x = width_centuries, 
        y = y_centuries, 
        orientation = "h"
    )

    Layout = go.Layout(
    title = "batsmen's performance",
    xaxis = dict(title = "No. of centuries"),
    yaxis = dict(title = "Batsmen")
    )

    fig = go.Figure(
        data = [trace],
        layout = Layout
    )

    fig.show()
```

# Question 9 

## Strike Rate vs Overs for top 10 Batsmen

```python
data = []
for batsman in data_deliveries["batsman"].value_counts().head(10).index:
  strike_rate_list = []
  df = data_deliveries[data_deliveries["batsman"] == batsman]
  df1 = df[["over", "batsman_runs"]]
  for over in range(1, 21):
    runs = df1[df1['over'] == over]['batsman_runs'].sum()
    balls = len(df1[df1['over'] == over])
    if balls != 0:
      strike_rate = (runs/balls)*100
      strike_rate_list.append(strike_rate)
    else:
      strike_rate_list.append(0)

  trace = go.Scatter(
      x = np.arange(1, 21),
      y = strike_rate_list,
      mode = 'markers',
      name = batsman
  )


  data.append(trace)
  

Layout = go.Layout(
    title = "strike rate per over vs overs for different batsmen",
    xaxis = dict(title = "overs"),
    yaxis = dict(title = "strike rate per over")
  )

fig = go.Figure(
    data = data,
    layout = Layout
)

fig.show()
```

# Question 10

## Team Comarison

```python
def compair_teams(team1, team2, get_bibar = True, get_percentage_barh = True, get_bibar_of_wins_against_each_other = True):

  if get_bibar:
    df = data_matches[["season", "winner"]]

    matches_won_team1 = []
    for year in sorted(df["season"].unique()):
      df1 = df[df["winner"] == team1]
      wins = len(df1[df1["season"] == year])
      matches_won_team1.append(wins)

    matches_won_team2 = []
    for year in sorted(df["season"].unique()):
      df1 = df[df["winner"] == team2]
      wins = len(df1[df1["season"] == year])
      matches_won_team2.append(wins)

    trace1 = go.Bar(
        x = np.array(sorted(df["season"].unique())),
        y = matches_won_team1,
        name = team1
    )

    trace2 = go.Bar(
        x = np.array(sorted(df["season"].unique())),
        y = matches_won_team2,
        name = team2
    )

    layout = go.Layout(
        title = f"{team1} vs {team2} seasonwise total wins data",
        xaxis = dict(title = "season",
                     tickmode = "linear"),
        yaxis = dict(title = "Total wins")
    )

    data = [trace1, trace2]

    fig = go.Figure(
        data = data,
        layout = layout
    )

    fig.show()


  if get_percentage_barh:
    win_percentage_team_comparison = []
    for team in [team1, team2]:
      matches_played = len(data_matches[data_matches["team1"] == team]) + len(data_matches[data_matches["team2"] == team])
      matches_won = len(data_matches[data_matches["winner"] == team])
      percentage_win = (matches_won / matches_played) * 100
      win_percentage_team_comparison.append(percentage_win)

    trace = go.Bar(
        x = win_percentage_team_comparison,
        y = [team1, team2], 
        orientation = "h"
    )
    
    layout = go.Layout(
        title = f"{team1} vs {team2} percentage wins",
        xaxis = dict(title = "percentage matches won till now"),
        yaxis = dict(title = "teams")
    )

    fig = go.Figure(
        data = [trace],
        layout = layout
    )

    fig.show()

  if get_bibar_of_wins_against_each_other:
    team1_wins = []
    team2_wins = []
    df = data_matches[((data_matches["team1"] == team1) & (data_matches["team2"] == team2)) | ((data_matches["team1"] == team2) & (data_matches["team2"] == team1))]
    df1 = df[["season", "winner"]]
    for year in sorted(df1["season"].unique()):
      win_by_team1 = list(df1[df1["season"] == year]["winner"]).count(team1)
      win_by_team2 = list(df1[df1["season"] == year]["winner"]).count(team2)

      team1_wins.append(win_by_team1)
      team2_wins.append(win_by_team2)

    trace1 = go.Bar(
        x = np.array(sorted(df1["season"].unique())),
        y = team1_wins,
        name = team1
    )

    trace2 = go.Bar(
        x = np.array(sorted(df1["season"].unique())),
        y = team2_wins,
        name = team2
    )

    data = [trace1, trace2]

    layout = go.Layout(
        title = f"{team1} vs {team2} seasonwise performance against each other",
        xaxis = dict(title = "season",
                     tickmode = "linear"),
        yaxis = dict(title = "No. of wins against each other")
    )

    fig = go.Figure(
        data = data,
        layout = layout
    )

    fig.show()
```

```python
compair_teams("Chennai Super Kings", "Mumbai Indians")
```
